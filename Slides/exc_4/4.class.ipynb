### Plotting

Python has multiple module for plotting.  
We will mainly work with `matplotlib` and some `seaborn`.  

[Cheatsheet](https://github.com/matplotlib/cheatsheets) for the matplotlib module

Lets try it out using a simple function:
\begin{equation*}
    f(x) = \sin(x)+0.05x^2
\end{equation*}

```python
import matplotlib.pyplot as plt
import numpy as np

# define the function
def f(x):
    fx = np.sin(x)+0.05*x**2 
    return fx

# generate some data
N = 100
x_vec = np.linspace(-10,10,N) # draw N=100 evenly spaced numbers between -10 and 10
```

```python
# First initialize canvas for plots.
fig = plt.figure(figsize = (10,7)) # From documentation: "Unique identifier for the figure"

# Add plot to canvas.
ax = fig.add_subplot(1,1,1) # From documentation: add_subplot(nrows, ncols, index, **kwargs)
ax.plot(x_vec, f(x_vec), ls = '--', color = 'red', label = 'function') # Choose plot method: .plot(), .hist(), .plot_surface() etc.
ax.plot(x_vec, -f(x_vec), color = 'green', label = 'negative function')
ax.plot(-2,-1.4,color = 'blue', marker = 'o', markersize = 10, label = 'some point')

# Customize plot
ax.grid(True)
ax.legend()
ax.set_xlabel('$x$')
ax.set_ylabel('$f(x)$')
ax.set_title('Awesome plot'); #; suppress' print
```

```python
# multiple graphs on same canvas
fig2 = plt.figure(figsize = (12,7))
ax1 = fig2.add_subplot(2,2,1)
ax1.plot(x_vec, f(x_vec))
ax1.set_title('First plot')

ax2 = fig2.add_subplot(2,2,4)
ax2.plot(np.cos(x_vec), (f(x_vec)))
ax2.set_title('Second plot');
```

### Lets optimize the function  
Finding the global optimum of the function can be done in multiple ways:


* We can loop through possible solutions and return the best one
    * Cumbersome and optimum not guaranteed!

* We can use the science python (**scipy**) module
    * We can use `optimize.minimize_scalar()` when optimizing just 1 variable
    * We can use `optimize.minimize()` when optimizing multiple variables - also works for just 1 variable
    * **Note:** you can maximize by minimizing the negative objective function 

```python
from scipy import optimize

# define initial guess
x_guess = 0

# objective function to optimize
obj = lambda x: f(x)

# optimize the objective function starting at x_guess
res = optimize.minimize(fun = obj,
                        x0 = x_guess,
                        method = 'Nelder-Mead')


# nice printout of optimization
print('-'*35)
print(res.message)
print('-'*35)
print('Using a numerical solver, the optimal values are:')
print(f'function value = {res.fun:.3f} with x = {res.x[0]:.3f}')
```

### Minimization methods
There are multiple minimization methods.
Generally they can be divided into two groups:
1. **gradient-based**
    * Relies on slope (*first derivative*) to take *steps*.  
    * Will take few steps (**fast**) but requires a *smooth looking function*.  
    * For example `SLSQP`, `BFGS` and `Newton-CG`  
2. **gradient-free**
    * Take **lots** of steps (**slow**), but performs well in *noisy functions*.  
    * For example `Nelder-Mead` which is a downhill method taking many steps based on improvement/non-improvement


`SLSQP` is neat when faced with **budget constraints** as the optimization can be bounded.
